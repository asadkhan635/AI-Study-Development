```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
df=pd.read_csv(r"D:\study bs maths first sem\application of ict\lab\ai_agent\data.csv")
print("data is loaded successfully")
```

```python
df.info()
```

```python
df.describe()
```

```python
df.info()
df.brand_map = df.groupby('brand')['Brand Ratio'].mean().to_dict()
```

```python
features=['Age','fuel','owner','km_driven','seller_type',"segment ratio",'transmission','Brand Ratio']
```

```python
x=df[features]
y=df["selling_price"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=RandomForestRegressor()
model.fit(x_train, y_train)
```

Check the results 

```python
print(f"training complete")
print(f"model accuracy: {model.score(x,y)*100:2f}%")
print(f"r2 score: {r2_score(y_test, model.predict(x_test))*100:2f}%")
print(f"mean absolute error: ",mean_absolute_error(y_test, model.predict(x_test)))
```

```python
age=int(input("Enter age of car in years: "))
owner=int(input("Enter number of previous owners (Test Drive car=0, first owner=1, second owner=2, Third owner=3, forth and more owner=4): "))
fuel=int(input("Enter fuel type ( CNG=1, LPG=0,Petrol=2, Diesel=3, Electric=4): "))
seller_type=int(input("Enter seller type (Individual=0, Dealer=1, Trustmark Dealer=2): "))
km_driven=int(input("Enter kilometers driven: "))
transmission=int(input("Enter transmission type (Manual=0, Automatic=1): "))
sub_catogory=input("Enter sub category  of car : ")
brand=input("Enter brand of car : ")

```

```python
segment_lookup=df.groupby('sub_category')["segment ratio"].mean().to_dict()

segment_ratio=segment_lookup.get(sub_catogory,0)
```

```python
brand_lookup=df.groupby('brand')["Brand Ratio"].mean().to_dict()

brand_ratio=brand_lookup.get(brand,0)
```
